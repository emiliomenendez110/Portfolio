# EXPLORATORY DATA ANALYSIS

```python
# Importying libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing data
df = pd.read_csv("Final_Data.csv", index_col=False)
```

```python
# Showing first five rows
df.head(5)
```

```python
# setting format for graphics
sns.set_theme(style='dark', palette='colorblind', font='Arial')
plt.figure(figsize=(25,10))
```

```python
# Count Plot Online Delivery
count_online_order = sns.countplot(x="online_order", data=df);
count_online_order.set(title='Online delivery')
for i in count_online_order.containers:
    count_online_order.bar_label(i)
```

```python
# Pie Chart Online Delivery
online, no_online = df["online_order"].value_counts()
pie_online_order = plt.pie([online, no_online], labels=["Yes","No"],
            colors=["green","blue"],
            pctdistance=1.25, labeldistance=.6,
            autopct='%1.2f%%')
plt.title('Online delivery')
plt.show()
```

A total of 27046 (65.66%) restaurants online deliver food and 14144 (34.34%) restaurants do not

Booking

```python
# Count Plot Booking
count_booking = sns.countplot(x="booking", data=df);
count_booking.set(title='Booking')
for i in count_booking.containers:
    count_booking.bar_label(i)
```

```python
# Pie Chart Booking
yes, no = df["booking"].value_counts()
pie_booking = plt.pie([yes, no], labels=["Yes","No"],
            colors=["green","blue"],
            pctdistance=1.25, labeldistance=.6,
            autopct='%1.2f%%')
plt.title('Booking')
plt.show()
```

Top 10 Locations

```python
# Bar Plot locations
locations = df["location"].value_counts()[:10].sort_values(ascending = False)
bar_locations = sns.barplot(x=locations,y=locations.index)
bar_locations.set(title= "Top 10 Locations")
for i in bar_locations.containers:
    bar_locations.bar_label(i)
```


Top 10 Rest Type


```python
# Bar restaurant type
restaurant_type_countplot = sns.countplot(y ='restaurant_type', data= df, order = df['restaurant_type'].value_counts().iloc[:10].index)
restaurant_type_countplot.set(title= "Restaurant Type")

for i in restaurant_type_countplot.containers:
    restaurant_type_countplot.bar_label(i)
```

Top 10 Food Type

```python
food_type = df["food_type"].value_counts()[:10].sort_values(ascending = False)
bar_food = sns.barplot(x=food_type,y=food_type.index)
bar_food.set(title= "Top 10 Food Type")
for i in bar_food.containers:
    bar_food.bar_label(i)
```

Services Type

```python
type = df["type"].value_counts().sort_values(ascending = False)
bar_type = sns.barplot(x=type,y=type.index)
bar_type.set(title= "Service Type")
for i in bar_type.containers:
    bar_type.bar_label(i)
```

Relation between Type and Rating


```python
plt.figure(figsize=(10,20))
rest_type_rate = sns.barplot(y="restaurant_type",x="rate",data=df)
rest_type_rate.set(title="Restaurant Type and Rating")
```

No. of restaurants in a Location

```python
#Bar plot Restaurants by location
plt.figure(figsize=(10,20))
restaurant_location = df.groupby("location")["restaurant_name"].nunique().sort_values(ascending=False)
bar_restaurant_location = sns.barplot(x=restaurant_location,y=restaurant_location.index)
bar_restaurant_location.set(title= "No. Restaurants by Location")
for i in bar_restaurant_location.containers:
    bar_restaurant_location.bar_label(i)
```

Average cost by City

```python
df.groupby("city")["cost"].mean()
```

```python
plt.figure(figsize=(10,20))
mean_restaurant_type = df.groupby("city")["cost"].mean().sort_values(ascending=False)
bar_mean_city = sns.barplot(x=mean_restaurant_type,y=mean_restaurant_type.index,data=df)
bar_mean_city.set(title="Average Cost by City")
for i in bar_mean_city.containers:
    bar_mean_city.bar_label(i)
```
