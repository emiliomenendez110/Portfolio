# DATA CLEANING

```python
# Importing data
df = pd.read_csv("Data.csv")
```

#### Eliminating columns that do not provide useful information
#### deleted columns:
    -url
    -address
    -phone
    -dish_liked
    -reviews_list
    -menu_items

```python

df.drop(columns=['url','address','phone','dish_liked','reviews_list','menu_item'],
        inplace=True)
```

```python
# Showing the columns
list(df.columns)
```

#### Renaming Columns

```python
# Creating a dictionary with the new column names
df.rename(columns={"name":"restaurant_name",
                   "online_order":"online_order",
                   "book_table":"booking",
                   "rate":"rate",
                   "votes":"votes",
                   "location":"location",
                   "rest_type":"restaurant_type",
                   "cuisines":"food_type",
                   "approx_cost(for two people)":"cost",
                   "listed_in(type)":"type",
                   "listed_in(city)":"city"}
          ,inplace = True)
```

```python
# Showing the columns
list(df.columns)
```

```python
# Null values
df.isnull().sum()
```

```python
# removing null values
df.dropna(axis=0, inplace=True)
```

#### Cleaning Individual Columns

```python
df[df.duplicated()].count().sum()
```

```python
list(df["restaurant_name"].unique())
```

```python
# Unique values from Online order column
df["online_order"].unique()
```

```python
# Unique values from Online order column
df["booking"].unique()
```

```python
# Modifying and creating new rating
df["rate"] = df["rate"].str.split("/", expand=True)[0]
```

```python
# Unique values rate column
df["rate"].unique()
```

```python
# Importing libraries
import numpy as np

# Become values "NEW" and "-" to nan
df["rate"].replace(["NEW", "-"], np.nan, inplace=True)

#converting the rating column to float
df["rate"] = df["rate"].astype(dtype="float")
```

```python
df["rate"].unique()
```

```python
# Unique values rate column
df["votes"].unique()
```

```python
# Unique values rate column
df["location"].unique()
```

```python
# Removing the comma from the cost column
df["cost"] = df["cost"].str.replace(",", "")

# Changing from object to integer
df["cost"] = df["cost"].astype(dtype="int64")
```

```python
# null values
df.isnull().sum()
```

```python
# removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# removing null values
df.dropna(axis=0, inplace=True)
```

```python
print("Rows :", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# Final Dataset
df.info()
```

```python
# Final Dataset
df.to_csv("Final_Data.csv", index=False)
```
